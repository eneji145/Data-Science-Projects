The target variable is RainTomorrow - so wherever this contains a null value we drop the entire row

```python
Cleaned_Data.loc[Cleaned_Data["month"] == 12, "Seasons"] = "Summer"
# using the loc function to assign December to Summer in the "Seasons" column
```

# MAKING USE OF HYPOTHESIS TESTING TO CONFIRM RELATIONSHIPS BETWEEN SOME OF THE COLUMNS

Making use of Chi2_contingency on columns that are of object datatype

```python
from scipy.stats import chi2_contingency as c2
```

Null Hypothesis - There is no correlation between the WindGustDirection and rain falling the next day

alpha value - 0.05

```python
ct = pd.crosstab(Cleaned_Data["WindGustDir"], Cleaned_Data["RainTomorrow"])
```

```python
c2(ct)
```

The pvalue (3.59 * 10 ^ -149) < alpha value (0.05)

Therefore we reject the null hypothesis. Meaning that there is indeed a correlation between the Wind Gust Direction and rain falling the next day

# TO KNOW IF THERE IS MAX TEMP AND MIN TEMP HAVE SIGNIFICANT EFFECT ON THE POSSIBILITY OF RAIN FALLING THE NEXT DAY

Using Ordinary Least Squares from statsmodels

```python
from statsmodels.api import OLS
```

Remember that the target variable "RainTomorrow" is of object datatype so it has to be converted to integer datatype for it to work just fine

```python
Cleaned_Data["RainTomorrow"].unique()
```

```python
Cleaned_Data.loc[Cleaned_Data["RainTomorrow"] == "Yes", "NextDayRain"] = 1
Cleaned_Data.loc[Cleaned_Data["RainTomorrow"] == "No", "NextDayRain"] = 0
```

Null Hypothesis - MaxTemp and MinTemp of the current day have no significant effect on Rain falling the next day

Alternate Hypothesis - Either MaxTemp or MinTemp of the current day have a significant effect on Rain falling the next day

alpha value - 0.05

```python
import statsmodels.api as sm

X = Cleaned_Data[["MaxTemp", "MinTemp"]] #independent variables

y = Cleaned_Data["NextDayRain"] # dependent variable

X = sm.add_constant(X) #adding an intercept

model = sm.OLS(y, X).fit()

```

```python
model.summary()
```

We can see that the p-values of the MaxTemp(0) and the MinTemp(0) are both lower than the alpha value(0.05). Meaning that we reject the null hypothesis. 

This means both MinTemp and MaxTemp of the current day have a significant effect on rain falling the next day

# DEVELOPING AN XGBOOST MODEL

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
```

From the above it is evident that the model performed very well in identifying the True Negatives, however it did not effectively classify the True positives

Making use of Neural Networks to make prediction

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
```

```python
cols = X.shape
```

```python
y = to_categorical(y)
```

```python
model = Sequential()

model.add(Dense(100, activation = "relu", input_shape = (cols[1],)))

model.add(Dense(100, activation = "relu"))

model.add(Dense(100, activation = "relu"))

model.add(Dense(2, activation = "softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer = "adam", loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
early_stopping = EarlyStopping(patience = 3, monitor = "val_accuracy")
```

```python
model.fit(X, y, validation_split = 0.2, callbacks = [early_stopping], batch_size = 32, epochs = 32)
```
